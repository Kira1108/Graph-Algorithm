## The power law distribution

$p(x) = Cx^{-\alpha} = \frac{C}{x^\alpha}$, $x \geq x_{min}$ (avoid division by 0)     
Power law distribution is only normalizable for $\alpha > 1$    
Normalizing constant $C = (\alpha - 1) x_{min}^{(\alpha -1)}$

**Power law PDF**
$p(x) = \frac{(\alpha - 1)}{x_{min}} (\frac {x}{x_{min}})^{-\alpha}$    

**Complementary CDF**
$\overline{F}(x) = \int_{x}^{\infty} p(x) dx = (\frac{x}{x_{min}})^{-(\alpha - 1)}$

*Both curve will be straight line on loglog scale*

- $p(x) = Cx^{-\alpha}, \overline{F}(x) = \overline{C}x^{-(\alpha - 1)}$    
- $\log p(x) = \log C - \alpha\log x$      
- $\log \overline{F}(x) = \log \overline{C} - (\alpha -1)\log x $

## Erdos Renny Graph

```
G = nx.erdos_renyi_graph(1000,0.2)
degrees = np.array([d for n, d in G.degree()])
mean_degree = np.mean([d for n, d in G.degree()])
print('Mean degree: {}'.format(mean_degree))

plt.hist(degrees)
plt.show()
```
