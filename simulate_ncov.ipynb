```
import networkx as nx
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from datetime import datetime 
pd.set_option('max_rows',1000)
%matplotlib inline
```

```
DATA_PATH = '/content/drive/My Drive/Network Codes/data_ncov'
```

```
n_suspect = 22
n_infected = 43
n_person = n_suspect + n_infected
n_location = 32
n_hospital = 10
n_traffic = 6
```

```
person = ['per_' + str(_) for _ in range(n_person)]
location = ['loc_' + str(_) for _ in range(n_location)]
hospital = ['hos_' + str(_) for _ in range(n_hospital)]
traffic = ['loc_' + str(_) for _ in range(n_traffic)]
```

```
from faker import Faker
fake = Faker('zh_CN')
```

```
symptoms = ['发烧','咳嗽', '乏力', '肌肉酸痛', '呕吐']
n_symptoms = [i +1 for i in range(len(symptoms))]

person_nodes = np.c_[person, 
      [fake.name() for _ in range(n_person)], 
      np.random.randint(0.3,100,size = n_person),
      np.random.choice(['男','女'], size = n_person, replace = True),
      ['infected'] *  n_person,
      ]

p_symptoms = []
for p in range(n_person):
    n_syms = np.random.choice(n_symptoms,p = [0.1,0.1,0.3,0.4,0.1])
    sym = np.random.choice([1,2,3,4,5],p=[0.4,0.2,0.2,0.1,0.1],size = n_syms,replace = False)
    syms = [symptoms[s-1] for s in sym]
    p_symptoms.append(syms)
```

```
inf_causes = ['近距离飞播传播','接触传播','气溶胶传播']
causes = np.random.choice(inf_causes, size = n_person)

p_locations = ['{}, {}'.format(fake.province(),fake.city()) for _ in range(n_person)]

infect_date = [fake.date_between_dates(date_start=datetime(2020,1,1), date_end=datetime(2020,2,19)) for _ in range(n_person)]
```

```
person = pd.DataFrame(np.c_[person_nodes, p_symptoms, causes,p_locations,infect_date],
                      columns = ['person_id','name','age','gender','epedemic','symptoms','infect_type','location','infect_date'])
```

```
location = pd.DataFrame([['loc_'+ str(_),fake.address()] for _ in range(n_location)],columns = ['loc_id','address'])

hospital = pd.DataFrame([('hos_'+str(_) ,fake.province() + np.random.choice(['军区总','人民', '发热','保健','陆军','西城','海淀']) + '医院') 
    for _ in range(n_hospital)],
    columns = ['hos_id','hospital'])

train = ['G' + str(np.random.randint(10,999))+'-'+str(np.random.randint(1,12))+'车厢' for _ in range(3)]
bus = [str(np.random.randint(1,999)) + '路' for _ in range(1)]
subway = [str(np.random.randint(1,10)) + '号线' for _ in range(1)]
airline = [''.join(np.random.choice([str(chr(i)) for i in range(65,91)],size = 2).tolist()) + 
           str(np.random.randint(1000,9999)) + '次' for _ in range(1)]

traffic = train+bus+subway + airline

traffic = pd.DataFrame(np.c_[np.arange(n_traffic),traffic],columns = ['traffic_id','traffic'])
traffic.traffic_id = 'traffic_' +  traffic.traffic_id
```

```
n_infection = 20
infect_rel = pd.DataFrame(np.random.choice(person.person_id,size = (n_infection,2), replace = True),
             columns = ['infector','infectee'])
infect_rel.to_csv(os.path.join(DATA_PATH,'infect_relation.csv'))
```

```
n_location_cooccur = 40
location_rel = [[np.random.choice(person.person_id), np.random.choice(location.loc_id)] 
                for _ in range(n_location_cooccur)]
pd.DataFrame(location_rel, columns = ['person_id','location_id']).to_csv(os.path.join(DATA_PATH,'location_rel.csv'))
```

```
n_hos_accepted = 30

infected_person = np.unique(np.r_[infect_rel.infector,infect_rel.infectee])
hos_rel = [[np.random.choice(infected_person),np.random.choice(hospital.hos_id)] for _ in range(n_hos_accepted)]
pd.DataFrame(hos_rel,columns = ['person_id','hospital_id']).to_csv(os.path.join(DATA_PATH,'hospital_rel.csv'))
```

```
n_co_vechle = 30
traffic_rel = [
               [np.random.choice(person.person_id),
                np.random.choice(traffic.traffic_id),
                fake.date_between_dates(date_start=datetime(2020,1,1), 
                        date_end=datetime(2020,2,19))] for _ in range(n_co_vechle)]
pd.DataFrame(traffic_rel, columns = ['person_id','traffic_id','date']).to_csv('traffic_rel.csv')
```

```
person.to_csv(os.path.join(DATA_PATH,'person.csv'))
location.to_csv(os.path.join(DATA_PATH,'location.csv'))
hospital.to_csv(os.path.join(DATA_PATH,'hospital.csv'))
traffic.to_csv(os.path.join(DATA_PATH,'traffic.csv'))
```
