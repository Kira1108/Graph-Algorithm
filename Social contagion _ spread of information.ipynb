*Population (Similar to epidemics)*

- Susceptible - individual who has not learn new information
- Infected - the spreader of the information
- Recovered - aware of the information, but not spread it


*Problem*    
- If the rumor reach high number of population
- Rate of infection spread

### Maki - Thomson Model

- Ignorants - people ignorant to rumor (do not know the rumor yet) $I$
- Spreaders - people who heard and actively spread the rumor $S$
- Stiflers - heard, but ceased spreading rumor $R$


*Rumor spread by pairwise contacts*

$\lambda$ rate of ignorant - speader contacts        
$\alpha$ rate of spreader - spreader &  spreader - stifler contacts   

*Process*     
$S + I \xrightarrow{\lambda} 2S$     
$S + R \xrightarrow{\alpha} 2R$       
$S + S \xrightarrow{\alpha} R + S$    


*Note: some details may change the equation*


*Then you have a differential equation system*   
You can not fully solve it, but only take a look at what happens when $t \to \infty$         


*Got a approximation like*    

$r_{\infty} = 1 - e^{- (1 + \lambda/\alpha)r_{\infty}}$

```
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.xkcd()
from IPython.core.pylabtools import figsize
```

```
figsize(22,12)
params = [(0,4,0.7),(0.4,0.6),(0.4,0.5),(0.4,0.4),(0.4,0.3),(0.4,0.2)]
r = np.arange(0,1,0.01)
for i,param in enumerate(params):
    lambd,alpha = param[0],param[1]
    plt.subplot(2,3,i+1)
    right = 1 - np.exp(-(1 + lambd / alpha) * r) 
    plt.plot(r,right,linewidth = 3)
    plt.plot(r,r,linewidth = 3)
    plt.title('Lambda: {}. Alpha: {}, lam/alp: {}'.format(round(lambd,3), round(alpha,3), round(lambd / alpha,3)))
plt.show()
```

**No rumor epidemic threshold, Non zero solution always exists**       
**Rumors always spread**

### Change of rumor dynamics

 *Rumor dynamics*
 
 - Whenever a spreader contacts an ignorant, the ignorant becomes a spreader at a rate $\lambda$
 - When a spreader contacts a stifler the initiating spreader becomes a stifler at a rate $\alpha$
 - <font color = 'darkred'>Spreader may also cease spreading a rumor spontaneously(without the need for any contact) at a rate $\delta$</font>

*Found*      
 - Homogeneous networks and random graphs have epidemic threshold:

 $\frac{\lambda}{\delta} >= \frac{\langle k \rangle}{\langle k^{2} \rangle}$

 - In scale free networks threshold vanishes as size growth
